```python
import os
from PyPDF2 import PdfReader

# Define the path to the folder containing PDFs
folder_path = "/content/drive/MyDrive/capstone/resumes/Resumes"

# Define the path to the output folder for text files
output_folder_path = "TextFiles"
os.makedirs(output_folder_path, exist_ok=True)

# Loop through each PDF file in the folder
for filename in os.listdir(folder_path):
    if filename.endswith(".pdf"):
        # Construct the full path to the PDF file
        pdf_file_path = os.path.join(folder_path, filename)

        # Create a PdfReader object
        reader = PdfReader(pdf_file_path)

        # Create a .txt file to store the extracted text
        output_text_file_path = os.path.join(output_folder_path, os.path.splitext(filename)[0] + "_extracted_text.txt")
        output_text_file = open(output_text_file_path, "w", encoding="utf-8")

        # Loop through each page and extract/store text
        for page in reader.pages:
            page_text = page.extract_text()

            # Write the extracted text to the output text file
            output_text_file.write(page_text)

        # Close the output text file
        output_text_file.close()
```

```python
import os
import spacy
import pandas as pd

# Load spaCy NLP model
nlp = spacy.load("en_core_web_sm")

# Define the path to the folder containing extracted text files
text_folder_path = "/content/TextFiles"

# Define a function to extract skills
def extract_skills(doc):
    # Customize this function based on your text patterns and skill extraction logic
    skills = []
    for token in doc:
        if token.pos_ == "NOUN":  # You might want to adjust this condition based on your data
            skills.append(token.text)
    return skills

# List to store skills data
skills_data = []

# Loop through each text file in the folder
for filename in os.listdir(text_folder_path):
    if filename.endswith("_extracted_text.txt"):
        # Construct the full path to the text file
        text_file_path = os.path.join(text_folder_path, filename)

        # Read the content of the text file
        with open(text_file_path, "r", encoding="utf-8") as text_file:
            text_content = text_file.read()

        # Process text using spaCy
        doc = nlp(text_content)

        # Extract skills from the processed text
        skills = extract_skills(doc)

        # Add data to the skills_data list
        skills_data.append({"Filename": filename, "Skills": ", ".join(skills)})

# Convert skills_data to a DataFrame
df = pd.DataFrame(skills_data)

# Save the DataFrame to a CSV file
output_csv_path = "extracted_skills.csv"
df.to_csv(output_csv_path, index=False, escapechar="\\")
```

```python
import os
import spacy
import pandas as pd

# Load spaCy NLP model
nlp = spacy.load("en_core_web_sm")

# Define the path to the folder containing text files
text_folder_path = "/content/TextFiles"

# Load the skills list from a text file
with open("output.txt", "r") as skills_file:
    skills_list = [line.strip().lower() for line in skills_file]

# Define a function to extract skills
def extract_skills(doc):
    skills = set()  # Use a set to avoid duplicates
    for token in doc:
        if token.text.lower() in skills_list:
            skills.add(token.text)
    return skills

# List to store skills data
skills_data = []

# Loop through each text file in the folder
for filename in os.listdir(text_folder_path):
    if filename.endswith(".txt"):
        # Construct the full path to the text file
        text_file_path = os.path.join(text_folder_path, filename)

        # Read the content of the text file
        with open(text_file_path, "r", encoding="utf-8") as text_file:
            text_content = text_file.read()

        # Process text using spaCy
        doc = nlp(text_content)

        # Extract skills from the processed text
        skills = extract_skills(doc)

        # Add data to the skills_data list
        skills_data.append({"Filename": filename, "Skills": ", ".join(skills)})

# Convert skills_data to a DataFrame
df = pd.DataFrame(skills_data)

# Save the DataFrame to a CSV file
output_csv_path = "extracted_skills.csv"
df.to_csv(output_csv_path, index=False)
```
